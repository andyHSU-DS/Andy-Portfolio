```python
import requests
import bs4
```

```python
website='https://pleagueofficial.com/stat-team'
obj=requests.get(website)#response
print(obj.text)
```

```python
soup=bs4.BeautifulSoup(obj.text,'lxml')
```

```python
head=[]
cols=soup.find('table',id='main-table')
columns=cols.find('thead',class_='bg-deepgray')
columns_text=columns.find_all('th')
for text in columns_text:
    print(text.text)
    head.append(text.text)
print('-'*50)  
print(head)
```

```python
import re
```

```python
alldata=[]
alldata.append(head)
body=soup.find('tbody')
find=re.compile(r'/D+')
teamdata=body.find_all('tr',class_=find.search(r'odd even'))
for team in teamdata:
    oneteam=[]
    teamname=team.find('th').a.text
    print(teamname)
    oneteam.append(teamname)
    others=team.find_all('td')
    for other in others:
        otherdata=other.text
        oneteam.append(otherdata)
    
    alldata.append(oneteam)
    
    
print(alldata)
```

```python
import csv
with open('pleaguedata.csv','w',newline='') as f1:
    a=csv.writer(f1)
    for y in alldata:
        a.writerow(y)
```
