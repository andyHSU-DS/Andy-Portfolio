```python
class Account:
    #condition 帳戶餘額
    #amount    扣款金額
    def __init__(self, name, condition, amount, stock_inventory):
        self.name            = name
        self.condition       = condition
        self.amount          = amount
        self.inventory       = stock_inventory
        #記錄庫存的買賣
        self.buyin_condition = []
        #記錄庫存的實際情況
        self.inventory_list  = [stock_inventory]
        print(self.buyin_condition)

    def saving_money(self):
        self.condition += 100000

    def buyin_stock(self,stockprice):
        if self.condition <= self.amount :
            print('金額不足')
            self.inventory += 0

        else:
            #算現在的錢購買多少股
            buy_in_amount = math.floor(self.amount / stockprice)
            print(f"買入:{buy_in_amount}")
            #放到buyin_condition內
            self.buyin_condition.append(buy_in_amount)
            #庫存新增
            self.inventory += buy_in_amount
            #帳戶餘額更新
            self.condition -= self.amount
        self.inventory_list.append(self.inventory)
        return buy_in_amount

    def sell_stock(self,stockprice):
        #庫存>0才能賣
        if self.inventory > 0 :
            print('賣出')
            random_amount = round(random.uniform(0.1,0.7),3)
            print(random_amount)
            #預設全賣
            selling_amount = math.floor(self.inventory * random_amount) * stockprice
            #帳戶餘額更新
            self.condition += selling_amount
            #放到buyin_condition內
            self.buyin_condition.append(-(math.floor(self.inventory * random_amount)))
            賣出的股票 = math.floor(self.inventory*0.8) * (-1)
            #將庫存更新
            self.inventory -=  math.floor(self.inventory * random_amount)
            self.inventory_list.append(self.inventory)
            return (賣出的股票)
        elif self.inventory == 0:
            #print('庫存不足')
            self.buyin_condition.append('無動作')
            self.inventory = 0 
            self.inventory_list.append(self.inventory)
            return '沒得賣'


    #單純為了讓self.inventory_list可以數量對齊，不然沒買沒賣的時候就不會新增記錄到inventory_list
    def 補nan(self):
        self.inventory_list.append(np.NaN)

    def call_condition(self):
        return (self.name, self.condition, self.amount, self.inventory, self.buyin_condition, self.inventory_list)

    def 結算(self,stockprice):
        self.condition += self.buyin_condition[-1] * stockprice
        print(self.condition)


    

    #Andy
    @classmethod
    def Andy(cls):
        return cls('Andy',1000000,50000,2000)
```

```python
Andy   = Account.Andy()
```

```python
recent_trend_跌 = []
recent_trend_漲 = []
TSLA['買賣']     = np.NaN
TSLA['數量']     = np.NaN

for x in range(len(TSLA)-1):
    if x % 10 == 0:
        #anyone.saving_money()
        Andy.saving_money()
    else:
        pass

    if TSLA.iloc[x+1,11] < TSLA.iloc[x,11]:
        recent_trend_跌.append('跌')
        if len(recent_trend_跌) == 3:
            #anyone.buyin_stock(TSLA.iloc[x+1,3])
            TSLA.iloc[x+1,-2] = '買'
            TSLA.iloc[x+1,-1] = Andy.buyin_stock(TSLA.iloc[x+1,3])
            recent_trend_跌.clear()
        else:
            Andy.補nan()
            recent_trend_漲.clear()
    else:
        recent_trend_漲.append('漲')
        if len(recent_trend_漲) == 15:
            #anyone.sell_stock(TSLA.iloc[x+1,3])
            TSLA.iloc[x+1,-2] = '賣'
            TSLA.iloc[x+1,-1] = Andy.sell_stock(TSLA.iloc[x+1,3])
            recent_trend_漲.clear()
        else:
            Andy.補nan()
            recent_trend_跌.clear()
TSLA['庫存量'] = Andy.call_condition()[-1]


#將np.NaN用前面的值補齊
TSLA['庫存量'] = TSLA['庫存量'].ffill()
            
```

```python
class Stock_Spe:
    def __init__(self,Ticker):
        self.Ticker                = Ticker
        self.stock_df              = yf.download(self.Ticker,period='5mo',interval='1d')
        self.stock_df['SMA_Close'] = self.stock_df['Close'].rolling(10).mean()
        self.stock_df['SMA_Open']  = self.stock_df['Open'].rolling(10).mean()
        self.stock_df['SMA_High']  = self.stock_df['High'].rolling(10).mean()
        self.stock_df['SMA_Low']   = self.stock_df['Low'].rolling(10).mean()

        self.stock_df['CMA_Close'] = self.stock_df['Close'].expanding().mean()
        self.stock_df['CMA_Open']  = self.stock_df['Open'].expanding().mean()
        self.stock_df['CMA_High']  = self.stock_df['High'].expanding().mean()
        self.stock_df['CMA_Low']   = self.stock_df['Low'].expanding().mean()

        self.stock_df['EMA']       = self.stock_df['High'].ewm(span=10).mean()
        return self.stock_df
```
